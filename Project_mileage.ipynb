# Remove Missing Values

```python
df = df.dropna()
```

```python
df.info()
```

# Data Visualization

```python
sns.pairplot(df, x_vars= ['displacement','weight','acceleration','mpg'], y_vars= ['mpg']);
```

# Define Target Variable y and Feature X

```python
df.columns
```

```python
y= df['mpg']
```

```python
y.shape
```

```python
x= df[['displacement', 'horsepower', 'weight', 'acceleration']]
```

```python
x.shape
```

```python
x
```

# Scaling Data

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss= StandardScaler()
```

```python
x = ss.fit_transform(x)
```

```python
x
```

```python
pd.DataFrame(x).describe()
```

**After Standardization Mean is Zero and Standard Devlation is One**

# Train test Split Data

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, train_size=0.7, random_state=2529)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

# Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
lr.intercept_
```

```python
lr.coef_
```

**Mileage=23.4-1.05Displacement-1.68Horsepower-4.10Weight-0.115Acceleration+error**

# Predict Test Data

```python
y_pred = lr.predict(x_test)
```

```python
y_pred
```

# Model Accuracy

```python
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, r2_score
```

```python
mean_absolute_error(y_test, y_pred)
```

```python
mean_absolute_percentage_error(y_test, y_pred)
```

```python
r2_score(y_test, y_pred)
```

# Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree=2, interaction_only=True, include_bias=False)
```

```python
x_train2 = poly.fit_transform(x_train)
```

```python
x_test2 = poly.fit_transform(x_test)
```

```python
lr.fit(x_train2, y_train)
```

```python
lr.intercept_
```

```python
lr.coef_
```
